```python
from math import log

# Build a cost dictionary, assuming Zipf's law and cost = -math.log(probability).
words = open("words-by-frequency.txt").read().split()
wordcost = dict((k, log((i+1)*log(len(words)))) for i,k in enumerate(words))
maxword = max(len(x) for x in words)

def infer_spaces(s):
    """Uses dynamic programming to infer the location of spaces in a string
    without spaces."""

    # Find the best match for the i first characters, assuming cost has
    # been built for the i-1 first characters.
    # Returns a pair (match_cost, match_length).
    def best_match(i):
        candidates = enumerate(reversed(cost[max(0, i-maxword):i]))
        return min((c + wordcost.get(s[i-k-1:i], 9e999), k+1) for k,c in candidates)

    # Build the cost array.
    cost = [0]
    for i in range(1,len(s)+1):
        c,k = best_match(i)
        cost.append(c)

    # Backtrack to recover the minimal-cost string.
    out = []
    i = len(s)
    while i>0:
        c,k = best_match(i)
        assert c == cost[i]
        out.append(s[i-k:i])
        i -= k

    return " ".join(reversed(out))
```

```python
def visualize_use(min, max, word_list, freq_dict):
    local = min
    
    while local <= max:
        if freq_dict.get(word_list[local]):
            freq_dict[word_list[local]] = freq_dict.get(word_list[local]) + 1
        else:
            freq_dict[word_list[local]] = 1
        local = local + 1
    
    return freq_dict


def find_freqs(test):
    
    global_freq = {}
    count = 0
    keyword = "immport"
    
    for count in range(len(test)):
        if test[count].lower() == keyword:
            global_freq = visualize_use(max(0, count - 30), min(len(test) - 1, count + 30), test, global_freq)
            count = count + 60
        else:
            count = count + 1
            
    return global_freq
            
```

```python
from textblob import Word 

def find_use(freq_list, article):
    
    cytokine = False
    registry = False
    
    data = False
    reuse = False
    
    fcs = False
    
    flock = False
    
    galaxy = False
    
    gene = False
    expression = False
    list_ = False
    
    cyto = False
    meta = False
    
    primary = False
    secondary = False
    analysis = False
    
    rimmport = False
    
    at_least_one  = False
    
    for word in freq_list:
        if "cytokine" == Word(word.lower()).singularize():
            cytokine = True
            at_least_one  = True
        if "registry" == Word(word.lower()).singularize():
            registry = True
            at_least_one  = True            
        if "data" == Word(word.lower()).singularize():
            data = True
            at_least_one  = True
        if "reuse" == word.lower():
            reuse = True
            at_least_one  = True
        if "fcs" == word.lower():
            fcs = True
            at_least_one  = True
        if "flock" == word.lower():
            flock = True
            at_least_one  = True
        if "galaxy" == word.lower():
            galaxy = True
            at_least_one  = True
        if "gene" == Word(word.lower()).singularize():
            gene = True
            at_least_one  = True
        if "expression" == Word(word.lower()).singularize():
            expression = True
            at_least_one  = True
        if "cyto" == word.lower():
            cyto = True
            at_least_one  = True
        if "meta" == word.lower():
            meta = True
            at_least_one  = True
        if "primary" == word.lower():
            primary = True
            at_least_one  = True
        if "secondary" == word.lower():
            secondary = True
            at_least_one  = True
        if "analysis" == word.lower():
            analysis = True
            at_least_one  = True
        if "rimmport" in word.lower():
            rimmport = True
            at_least_one  = True

    print("DATA FOR ARTICLE " + article + "\n")

    if cytokine and registry:
        print("Uses Cytokine Registry\n")
    if data and reuse:
        print("Uses Data Reuse\n")
    if fcs:
        print("Uses FCSTrans\n")
    if galaxy:
        print("Uses Galaxy\n")
    if gene and expression:
        print("Uses Gene Expression\n")
    if gene and list_:
        print("Uses Gene List\n")
    if cyto and meta:
        print("Uses MetaCyto\n")
    if primary and analysis:
        print("Uses Primary Analysis\n")
    if secondary and analysis:
        print("Uses Secondary Analysis\n")
    if rimmport:
        print("Uses RImmPort\n")
   
    if not at_least_one:
        return 1
    else:
        return 0
        
```

```python
def make_term_global(freq_list):
    
    global_dict = {}
    for article in freq_list:
        for word in freq_list[article]:
            if global_dict.get(word):
                global_dict[word] = global_dict[word] + freq_list[article][word]
            else:
                global_dict[word] = freq_list[article][word]
    return global_dict
```

```python
def make_df(freq_list):
    
    global_dict = {}
    for article in freq_list:
        for word in freq_list[article]:
            if global_dict.get(word):
                global_dict[word] = global_dict[word] + 1
            else:
                global_dict[word] = 1
    return global_dict
```
